```python
#Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing the dataset
dataset = pd.read_csv(r'C:\Users\LANRE\Desktop\ML-week3\ML-Week3-master\Salary_Data.csv')

dataset
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X,y)
```

```python
regressor.intercept_
```

```python
regressor.coef_
```

```python
print('The equation of linear regression is; yhat = 9449.96X + 25792.2')
```

```python
salarypred = regressor.predict(X)
```

```python
dataset['salary_pred']= salarypred
```

```python
dataset
```

```python
#Visualizing the result
plt.scatter(X,y, color = 'green')
plt.plot(X,salarypred, color = 'blue')
plt.title('Years of experience vs Salary')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn import metrics
print('Mean absolute error:', metrics.mean_absolute_error(y,salarypred))
print('Mean Squared Error:', metrics.mean_squared_error(y, salarypred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y, salarypred)))
```
